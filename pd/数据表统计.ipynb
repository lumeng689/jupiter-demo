```python
df = pd.read_csv("~/Downloads/query2.csv")
df["日期"] = df["日期"].astype(str)
df["占比"] = df["热点数"]/ df["总数"]
print(df[["日期","占比"]])
```

```python
df.plot(x='日期', y=['热点数'], kind='line')
plt.rcParams['font.family'] = 'Heiti TC'  # 替换为你选择的字体
plt.show()
```
